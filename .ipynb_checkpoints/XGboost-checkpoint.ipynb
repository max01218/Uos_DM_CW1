```python
import pandas as pd
import numpy as np
# load the provided data
test_df = pd.read_csv('data/dengue_test_final_clean.csv')
train_df = pd.read_csv('data/dengue_train_final_clean.csv')
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
def fill_missing_with_neighbors_avg(series):
    """
    使用前后最近的三个非空值的平均值填补 NaN。
    """
    series_filled = series.copy()
    nan_indices = np.where(series.isna())[0]  # 找到所有 NaN 的索引

    for idx in nan_indices:
        # 取前3个最近的非空值
        prev_values = series.iloc[max(0, idx-3):idx].dropna().values
        # 取后3个最近的非空值
        next_values = series.iloc[idx+1:idx+4].dropna().values

        # 合并前后数据
        valid_values = np.concatenate([prev_values, next_values])

        if len(valid_values) > 0:
            series_filled.iloc[idx] = np.mean(valid_values)  # 计算均值填充

    return series_filled

# 处理 train_df
for column in train_df.columns:
    train_df[column] = fill_missing_with_neighbors_avg(train_df[column])

# 处理 test_df
for column in test_df.columns:
    test_df[column] = fill_missing_with_neighbors_avg(test_df[column])
```

# optuna hyperperemeter

```python
import pandas as pd
import numpy as np
import optuna
from xgboost import XGBRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold

# 假設原始的 X_train, X_test, y_train, y_test 全都是 pandas DataFrame/Series
# (若不是，請自行修改為相對應的格式)

# 1. 建立並套用 MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# 2) 將縮放後的 numpy array 再轉回 DataFrame
X_train_scaled = pd.DataFrame(X_train_sc, 
                              index=X_train.index, 
                              columns=X_train.columns)

X_test_scaled = pd.DataFrame(X_test_sc, 
                             index=X_test.index, 
                             columns=X_test.columns)

# 2. 定義 Optuna 目標函式
def objective(trial):
    # 擴大超參數空間
    n_estimators = trial.suggest_categorical('n_estimators', [100, 300, 500, 700, 1000])
    max_depth = trial.suggest_categorical('max_depth', [3, 5, 7, 9, 11])
    learning_rate = trial.suggest_float('learning_rate', 0.001, 0.2, log=True)
    subsample = trial.suggest_float('subsample', 0.5, 0.9, step=0.1)
    colsample_bytree = trial.suggest_float('colsample_bytree', 0.5, 1.0, step=0.1)
    min_child_weight = trial.suggest_categorical('min_child_weight', [1, 3, 5, 7, 10])
    gamma = trial.suggest_float('gamma', 0.0, 0.5, step=0.05)
    reg_alpha = trial.suggest_float('reg_alpha', 1e-3, 10.0, log=True)  # 下限 1e-3
    reg_lambda = trial.suggest_float('reg_lambda', 1e-3, 20.0, log=True)

    # 建立 XGBoost 模型（若有 GPU，則可使用 'cuda' 作為加速）
    model = XGBRegressor(
        objective='reg:squarederror',
        random_state=42,
        tree_method='hist',
        device='cuda',    # 如果沒有 GPU，請改為 'cpu' 或移除此參數
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        min_child_weight=min_child_weight,
        gamma=gamma,
        reg_alpha=reg_alpha,
        reg_lambda=reg_lambda,
        verbosity=0
    )
    
    # 3. 使用 KFold 交叉驗證（以 5 折為例）
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    mae_list = []
    
    # 因為我們把 X_train_scaled, y_train 都保留在 DataFrame/Series 中，
    # 故能使用 .iloc 取對應的資料列
    for train_idx, valid_idx in cv.split(X_train_scaled):
        X_tr, X_val = X_train_scaled.iloc[train_idx], X_train_scaled.iloc[valid_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[valid_idx]

        model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], verbose=False)
        preds = model.predict(X_val)
        mae = mean_absolute_error(y_val, preds)
        mae_list.append(mae)
        
    return np.mean(mae_list)

# 4. 建立 Optuna study，並使用 HyperbandPruner 進行較有效率的剪枝
study = optuna.create_study(
    direction='minimize',
    pruner=optuna.pruners.HyperbandPruner()
)

# 5. 開始超參數搜尋
study.optimize(objective, n_trials=100)

# 6. 輸出最佳結果與參數
print("最佳 MAE:", study.best_value)
print("最佳參數:")
for key, value in study.best_trial.params.items():
    print(f"  {key}: {value}")

best_params = study.best_trial.params
```

```python
import numpy as np
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import RandomizedSearchCV, HalvingGridSearchCV
from xgboost import XGBRegressor

# 設置 XGBoost 回歸模型
clf1 = XGBRegressor(objective='reg:squarederror',
                    random_state=42,
                    tree_method="hist", 
                    device="cuda")

# 更新參數範圍
param_dist = {
    'n_estimators': [100, 300, 500, 700], 
    'max_depth': [5, 7, 9, 11],  
    'learning_rate': np.linspace(0.005, 0.1, 10),  
    'subsample': np.linspace(0.5, 0.9, 5),  
    'colsample_bytree': np.linspace(0.7, 1.0, 5),  
    'min_child_weight': [1, 3, 5, 7],  
    'gamma': np.linspace(0.01, 0.3, 5),  
    'reg_alpha': [0.01, 0.1, 1, 10],  
    'reg_lambda': [0.1, 1, 10, 20]  
}

# 先進行快速搜索
random_search = RandomizedSearchCV(
    estimator=clf1, 
    param_distributions=param_dist,
    n_iter=50,  
    cv=5, 
    scoring='neg_mean_squared_error',
    verbose=1, 
    n_jobs=-1,
    random_state=42
)

# 進行訓練
random_search.fit(X_train, y_train)

# 獲取最佳參數
best_params = random_search.best_params_
print("RandomizedSearch 最佳參數:", best_params)

# 進一步細化搜索
fine_tune_param_dist = {
    'n_estimators': [best_params['n_estimators'] - 100, best_params['n_estimators'], best_params['n_estimators'] + 100],
    'max_depth': [max(best_params['max_depth'] - 1, 3), best_params['max_depth'], best_params['max_depth'] + 1],
    'learning_rate': np.linspace(max(best_params['learning_rate'] / 2, 0.005), min(best_params['learning_rate'] * 2, 0.2), 5),
    'subsample': np.linspace(best_params['subsample'] - 0.1, best_params['subsample'] + 0.1, 5),
    'colsample_bytree': np.linspace(best_params['colsample_bytree'] - 0.1, best_params['colsample_bytree'] + 0.1, 5),
    'min_child_weight': [max(best_params['min_child_weight'] - 1, 1), best_params['min_child_weight'], best_params['min_child_weight'] + 1],
    'gamma': [max(best_params['gamma'] - 0.05, 0), best_params['gamma'], best_params['gamma'] + 0.05],
    'reg_alpha': [max(best_params['reg_alpha'] / 2, 0.01), best_params['reg_alpha'], best_params['reg_alpha'] * 2],
    'reg_lambda': [max(best_params['reg_lambda'] / 2, 0.1), best_params['reg_lambda'], best_params['reg_lambda'] * 2]
}

# 進行細緻搜索
halving_search = HalvingGridSearchCV(
    estimator=clf1, 
    param_grid=fine_tune_param_dist,
    cv=3, 
    scoring='neg_mean_squared_error',
    verbose=1, 
    n_jobs=-1,
    factor=2  # 每次縮小2倍範圍
)

halving_search.fit(X_train, y_train)
best_params = halving_search.best_params_
print("最佳參數:", best_params)
```

```python
from xgboost import XGBRegressor

xgb_model_sj = XGBRegressor(**best_params)

xgb_model_sj.fit(X_train, y_train)
from sklearn.metrics import mean_absolute_error, mean_squared_error

predictions = xgb_model_sj.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
print(mae)

import matplotlib.pyplot as plt
# 視覺化預測結果
plt.figure(figsize=(12, 6))
plt.plot(y_test.to_numpy(), label="Actual Cases", color='blue')
plt.plot(predictions, label="Predicted Cases", color='red', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Total Cases")
plt.title("Xgboost Prediction Results(sj)")
plt.legend()
plt.show()
```

# optuna

```python
import numpy as np
import optuna
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold

def objective(trial):
    # 擴大超參數空間，部分參數改用對數取樣
    n_estimators = trial.suggest_categorical('n_estimators', [100, 300, 500, 700, 1000])
    max_depth = trial.suggest_int('max_depth', 3, 12, step=1)  # 從 3 到 12
    learning_rate = trial.suggest_float('learning_rate', 1e-3, 0.2, log=True)  # 對數取樣
    subsample = trial.suggest_float('subsample', 0.4, 1.0, step=0.1)
    colsample_bytree = trial.suggest_float('colsample_bytree', 0.4, 1.0, step=0.1)
    min_child_weight = trial.suggest_int('min_child_weight', 1, 10, step=1)
    gamma = trial.suggest_float('gamma', 1e-3, 1.0, log=True)  # 對數取樣
    reg_alpha = trial.suggest_float('reg_alpha', 1e-3, 10, log=True)  # 對數取樣
    reg_lambda = trial.suggest_float('reg_lambda', 1e-3, 20, log=True)  # 對數取樣

    # 建立 XGBoost 模型（使用 GPU 加速）
    model = XGBRegressor(
        objective='reg:squarederror',
        random_state=42,
        tree_method='hist',
        device='cuda',
        n_estimators=n_estimators,
        max_depth=max_depth,
        learning_rate=learning_rate,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        min_child_weight=min_child_weight,
        gamma=gamma,
        reg_alpha=reg_alpha,
        reg_lambda=reg_lambda,
        verbosity=0
    )
    
    # 使用 KFold 進行交叉驗證（5 折為例）
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    mae_list = []
    
    for train_idx, valid_idx in cv.split(X_train):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[valid_idx]
        
        model.fit(X_tr, y_tr, eval_set=[(X_val, y_val)], verbose=False)
        
        preds = model.predict(X_val)
        mae = mean_absolute_error(y_val, preds)
        mae_list.append(mae)
    
    return np.mean(mae_list)

# 使用 Hyperband 剪枝策略
study = optuna.create_study(direction='minimize', pruner=optuna.pruners.HyperbandPruner())
study.optimize(objective, n_trials=200)  # 視需求可加大 n_trials

print("最佳 MAE:", study.best_value)
print("最佳參數:")
for key, value in study.best_trial.params.items():
    print(f"  {key}: {value}")

best_params = study.best_trial.params
```

# HalvingGridSearchCV

```python
import numpy as np
from sklearn.model_selection import HalvingGridSearchCV
from xgboost import XGBRegressor

# 設置 XGBoost 回歸模型
clf2 = XGBRegressor(objective='reg:squarederror', random_state=42)

# 更新參數範圍
param_dist = {
    'n_estimators': range(150, 250, 10),  
    'max_depth': range(5, 15, 2),  
    'learning_rate': np.linspace(0.01, 0.2, 10),  
    'subsample': np.linspace(0.6, 0.9, 5),  
    'colsample_bytree': np.linspace(0.5, 0.9, 5),  
    'min_child_weight': range(1, 4, 1),  
    'gamma': np.linspace(0, 0.2, 5)
}

# 進行 Halving Grid Search
halving_search = HalvingGridSearchCV(
    clf2, param_dist,
    factor=2,  # 每輪減少一半的候選參數
    cv=3,
    scoring="neg_mean_squared_error",
    verbose=1,
    n_jobs=-1
)
halving_search.fit(X_train, y_train)
best_params = halving_search.best_params_
print("最佳參數:", best_params)
```
