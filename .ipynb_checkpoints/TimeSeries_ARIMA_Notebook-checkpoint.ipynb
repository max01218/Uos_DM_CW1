```python
def preprocess_data(data_path, labels_path=None):
    # load data and set index to city, year, weekofyear
    df = pd.read_csv(data_path, index_col=[0, 1, 2])
    
    # select features we want
    features = ['week_start_date','reanalysis_specific_humidity_g_per_kg', 'reanalysis_dew_point_temp_k', 
            'station_avg_temp_c', 'station_min_temp_c']
    df = df[features]
    
    # fill missing values
    df.fillna(method='ffill', inplace=True)

    # add labels to dataframe
    if labels_path:
        labels = pd.read_csv(labels_path, index_col=[0, 1, 2])
        df = df.join(labels)
    
    # separate san juan and iquitos
    return df
```

```python
from sklearn.preprocessing import LabelEncoder
def DF_change(df):
    df = df.reset_index()
    le = LabelEncoder()
    df['city'] = le.fit_transform(df['city'])
    df['week_start_date'] = pd.to_datetime(df['week_start_date'])  # Convert to datetime format
    df['month'] = df['week_start_date'].dt.month
    df['day'] = df['week_start_date'].dt.day
    df = df.drop(columns = 'week_start_date' )
    return df
df = preprocess_data('data/dengue_features_train.csv',
                                    labels_path="data/dengue_labels_train.csv")
df = DF_change(df)
```

```python
df.shape
```

```python
#split
train_df = df.head(1000)
test_df = df.tail(df.shape[0] - 1000) #936-800 = 136

# iq_train_subtrain = iq_train.head(400)
# iq_train_subtest = iq_train.tail(iq_train.shape[0] - 400)
```

```python
# å¿…è¦çš„å°Žå…¥
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima

# ç¢ºä¿æ•¸æ“šæŒ‰ç…§æ™‚é–“é †åºæŽ’åˆ—
train_df = train_df.sort_values(by=["year", "weekofyear"])
test_df = test_df.sort_values(by=["year", "weekofyear"])

# é¸æ“‡è¦é æ¸¬çš„ç›®æ¨™è®Šæ•¸ï¼ˆARIMA åªèƒ½ç”¨å–®è®Šé‡ï¼‰
target = 'total_cases'
y_train = train_df[target].values
y_test = test_df[target].values

# è‡ªå‹•é¸æ“‡æœ€ä½³ ARIMA åƒæ•¸ (p, d, q)
auto_arima_model = auto_arima(y_train, seasonal=False, stepwise=True, suppress_warnings=True)
p, d, q = auto_arima_model.order
print(f"æœ€ä½³ ARIMA åƒæ•¸: (p={p}, d={d}, q={q})")

# è¨“ç·´ ARIMA æ¨¡åž‹
arima_model = ARIMA(y_train, order=(p, d, q))
arima_fit = arima_model.fit()

# é æ¸¬æœªä¾†æ•¸æ“šé»ž
y_pred = arima_fit.forecast(steps=len(y_test))

# è¦–è¦ºåŒ–é æ¸¬çµæžœ
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_train)), y_train, label="Training Data", color='blue')
plt.plot(range(len(y_train), len(y_train) + len(y_test)), y_test, label="Actual Cases", color='green')
plt.plot(range(len(y_train), len(y_train) + len(y_test)), y_pred, label="ARIMA Predicted Cases", color='red', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Total Cases")
plt.title("ARIMA Prediction Results")
plt.legend()
plt.grid(True)
plt.show()
```
