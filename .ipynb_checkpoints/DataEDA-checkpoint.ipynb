```python
%matplotlib inline
from __future__ import print_function
from __future__ import division

import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
import statsmodels.api as sm

# just for the sake of this blog post!
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd

# 加載數據
features_path = "data/final_dengue_features_train.csv"
labels_path = "data/final_dengue_labels_train.csv"

df_features = pd.read_csv(features_path)
df_labels = pd.read_csv(labels_path)

# 顯示數據的基本信息
df_features.info(), df_labels.info()

# 顯示前幾行數據
df_features.head(), df_labels.head()
```

```python
# 合併數據集
df = pd.merge(df_features, df_labels, on=["city", "year", "weekofyear"], how="left")
```

```python
df
```

```python
# 為每個特徵與 total_cases 進行對比繪圖
for column in df.columns:
    if column != "total_cases":  # 避免 total_cases 和自身對比
        plt.figure(figsize=(12, 5))
        plt.plot(df["week_start_date"], df[column], label=column, color="blue", alpha=0.7)
        plt.plot(df["week_start_date"], df["total_cases"], label="total_cases", color="red", linestyle="dashed", alpha=0.7)
        
        plt.xlabel("Time")
        plt.ylabel("Value")
        plt.title(f"Trend of {column} vs. total_cases Over Time")
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 确保 week_start_date 是时间格式
df["week_start_date"] = pd.to_datetime(df["week_start_date"])

# 按月聚合
df_monthly = df.resample("M", on="week_start_date").sum()

# 找出峰值（尖端）
peak_indices = df_monthly["total_cases"].nlargest(5).index  # 取前 5 个峰值
peak_values = df_monthly.loc[peak_indices, "total_cases"]

# 绘制趋势图
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly["total_cases"], marker="o", linestyle="-", label="Monthly Cases")

# 标注峰值
for date, value in zip(peak_indices, peak_values):
    plt.annotate(f"{date.strftime('%Y-%m')}\n{value}", 
                 xy=(date, value), 
                 xytext=(date, value + 200),  # 文字位置稍微上移
                 ha="center", fontsize=10, 
                 arrowprops=dict(arrowstyle="->", color="red"))

# 图表美化
plt.xlabel("Time (Month)")
plt.ylabel("Number of Cases")
plt.title("Monthly Disease Cases Trend with Peak Labels")
plt.legend()
plt.grid(True)

# 显示图表
plt.show()
```

```python
# 確保 week_start_date 是 datetime 格式
df["week_start_date"] = pd.to_datetime(df["week_start_date"])

# 設置時間索引
df.set_index("week_start_date", inplace=True)

# 按季度重採樣 ndvi_ne
df_quarterly_ndvi = df["ndvi_ne"].resample("M").mean()  # 使用平均值來平滑趨勢

# 繪製季度趨勢圖
plt.figure(figsize=(12, 6))
plt.plot(df_quarterly_ndvi.index, df_quarterly_ndvi, marker="o", linestyle="-", label="Quarterly NDVI_NE", color="blue")

plt.xlabel("Time (Quarter)")
plt.ylabel("NDVI_NE Value")
plt.title("Quarterly NDVI_NE Trend")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_yearly = df.resample("Y", on="week_start_date").sum()

plt.figure(figsize=(12, 6))
plt.plot(df_yearly.index, df_yearly["total_cases"], marker="^", linestyle="-", label="Yearly Cases", color="red")

plt.xlabel("Time (Year)")
plt.ylabel("Number of Cases")
plt.title("Yearly Disease Cases Trend")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 添加月份和季度信息
df["month"] = df["week_start_date"].dt.month
df["quarter"] = df["week_start_date"].dt.quarter

# 按月份计算平均发病数，查看是否有季节性变化
monthly_avg_cases = df.groupby("month")["total_cases"].mean()

# 绘制趋势图
plt.figure(figsize=(10, 5))
plt.plot(monthly_avg_cases.index, monthly_avg_cases.values, marker="o", linestyle="-")

plt.xlabel("Month")
plt.ylabel("Average Number of Cases")
plt.title("Seasonal Pattern: Monthly Average Dengue Cases")
plt.xticks(range(1, 13), [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])  # 设置月份标签
plt.grid(True)
plt.show()
```

```python
# 按城市分开计算每月平均病例数
df["month"] = df["week_start_date"].dt.month
city_monthly_avg = df.groupby(["city", "month"])["total_cases"].mean().unstack(level=0)

# 绘制不同城市的趋势图
plt.figure(figsize=(12, 6))

for city in city_monthly_avg.columns:
    plt.plot(city_monthly_avg.index, city_monthly_avg[city], marker="o", linestyle="-", label=city)

# 图表美化
plt.xlabel("Month")
plt.ylabel("Average Number of Cases")
plt.title("Seasonal Dengue Cases Trend by City")
plt.xticks(range(1, 13), [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])
plt.legend(title="City")
plt.grid(True)

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 选择相关的环境因素列
env_factors = ["reanalysis_avg_temp_k", "reanalysis_precip_amt_kg_per_m2", "reanalysis_specific_humidity_g_per_kg"]

# 计算环境因素与 total_cases 的相关性
correlation_matrix = df[env_factors + ["total_cases"]].corr()

# 绘制热力图
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Environmental Factors and Dengue Cases")
plt.show()

# 绘制散点图以探索关系
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, factor in enumerate(env_factors):
    sns.scatterplot(data=df, x=factor, y="total_cases", alpha=0.5, ax=axes[i])
    axes[i].set_title(f"Total Cases vs {factor}")
    axes[i].set_xlabel(factor)
    axes[i].set_ylabel("Total Cases")

plt.tight_layout()
plt.show()
```
