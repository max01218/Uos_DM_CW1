```python
import pandas as pd
import numpy as np
# load the provided data
test_df = pd.read_csv('data/dengue_test_final_clean.csv')
train_df = pd.read_csv('data/dengue_train_final_clean.csv')
```

```python
# 提取 total_cases 時間序列數據
total_cases = train_df["total_cases"]
```

```python
alpha = 0.3  # 平滑係數
ewma_cases = total_cases.ewm(alpha=alpha).mean()
```

```python
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
# 可視化不同的平滑方法
plt.figure(figsize=(12, 6))
#plt.plot(total_cases, label="Original Total Cases", color="blue", alpha=0.5)
plt.plot(ewma_cases, label="EWMA Smoothed Cases", linestyle="dashed", color="red")


plt.xlabel("Time (Index)")
plt.ylabel("Total Cases")
plt.title("Smoothing Methods for Total Dengue Cases")
plt.legend()
plt.show()
```

```python
train_df["total_cases"] = ewma_cases
```
